```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv('news_1_mio_cleaned.csv')
df = df.drop(['Unnamed: 0'], axis = 1)

df['meta_keywords'] = df['meta_keywords'].apply(lambda x : str(x))
df
```

```python
df1 = pd.DataFrame(df.meta_keywords.str.split(',').tolist(), index=df.id).stack()
df1 = df1.reset_index([0, 'id'])
df1.columns = ['id', 'meta_keywords']

df1.replace('\W', "" , regex=True, inplace = True)
df1.replace(r'^\s*$', np.nan, regex=True, inplace = True)
df1.fillna(value='Unknown', inplace=True)

df1
```

```python
#keywords_table
df2 = df1['meta_keywords']
df2.drop_duplicates(keep ="first", inplace=True)
df2 = df2.to_frame()
df2.insert(0, 'keyword_id', range(100, 100 + len(df2)))
Keywords_table = df2
Keywords_table
```

```python
#tags
my_dict = pd.Series(df2.keyword_id.values, index = df2.meta_keywords).to_dict()
tags = df1
tags = df1.replace({"meta_keywords": my_dict})
tags.rename(columns={'id': 'article_id', 'meta_keywords': 'keyword_id'}, inplace = True)

tags
```

```python
#export to csv
Keywords_table.to_csv('keyword_table.csv', index = False)
tags.to_csv('tags_table.csv', index = False)
```
